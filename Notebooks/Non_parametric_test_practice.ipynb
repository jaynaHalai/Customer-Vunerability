# Sign Rank Test

```python
df['difference'] = df['After'] - df['Before']
```

```python
df
```

Ho: Median difference=0

H1: Median difference!= 0

```python
Postive = (df['difference'] > 0 ).sum()
Negative  = (df['difference'] < 0).sum()

print (f'Postive: {Postive}')
print(f'Negative: {Negative}')
```

```python
n= Postive + Negative
n
```

```python
Results = binomtest(k=Postive,n=n,p=0.5,alternative='two-sided')
print(Results)
```

```python
Results.pvalue
```

### conclusion 
Considering that the Pvalue > 0.05 we fail to reject the null hypothesis stating that the Median difference=0
